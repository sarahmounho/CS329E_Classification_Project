## prelims

removed the days columns and ventilator weaning/ventilator free days from the data to keep from overfitting

```python
data_pre = intake_data()
data_x = data_pre.drop(["death","days.1","ventilator weaning = 1", "VFD ","days"], axis=1)
data_y = data_pre["death"]
```

```python
data_x
```

## model selection

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np


scaler = StandardScaler()
perceptron = MLPClassifier()

steps = [("scaler",scaler), ("nn",perceptron)]
perceptron_pipeline = Pipeline(steps=steps)

params_grid = {
    "nn__hidden_layer_sizes":list((a,) for a in range(10,101,10)),
    "nn__activation":["logistic","tanh","relu"],
    "nn__solver":["lbfgs","adam"]
}

grid_search = GridSearchCV(perceptron_pipeline,params_grid)

results = cross_val_score(grid_search, data_x, data_y, cv=10)
```

## results

the cross_val_score

```python
print(np.mean(results))
```

the confusion matrix

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
y_pred = cross_val_predict(grid_search, data_x, data_y, cv=10)
conf_mat = confusion_matrix(data_y, y_pred)
```
